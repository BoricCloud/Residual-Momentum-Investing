```python
import pandas as pd
import numpy as np
```

```python
###############DATI FACTSET
```

```python
#lettura dataframe
SP_full = pd.read_csv('SP_full.csv').set_index('data')
EN_full = pd.read_csv('EN_full.csv').set_index('data')
CSI_full = pd.read_csv('CSI_full.csv').set_index('data')
```

```python
#elimino colonna indice
del SP_full['Unnamed: 0']
```

```python
del CSI_full['Unnamed: 0']
```

```python
del EN_full['Unnamed: 0']
```

```python
SP_full.head()
```

```python
#creazione rendimenti percentuali dei titoli rispetto al mese precedente
rendimento_SP = SP_full.pct_change()
rendimento_EN = EN_full.pct_change()
rendimento_CSI = CSI_full.pct_change()

rendimento_SP.head()
```

```python
#conversione del prezzo in scala logaritmica 
SP_log = SP_full.applymap(np.log10)
EN_log = EN_full.applymap(np.log10)
CSI_log = CSI_full.applymap(np.log10)

SP_log.head()
```

```python
#calcolo dei rendimenti logaritmici rispetto al mese precedente con eliminazione prima riga nulla
SP_rendimenti_log = SP_log.diff()[1:]
EN_rendimenti_log = EN_log.diff()[1:]
CSI_rendimenti_log = CSI_log.diff()[1:]

SP_rendimenti_log.head()
```

```python
###############ESEMPIO PRIMO INVESTIMENTO MAGGIO 2019
#creazione dataset con i primi 12 mesi per regressione
df_esempio = SP_rendimenti_log[0:12]
df_esempio
```

```python
#regressione dei rendimenti logaritmici mensili di tutti i titoli rispetto al rendimento logaritmico dell'indice

import statsmodels.api as sm #libreria per regressione

resids = {}
for column in df_esempio:
    
    #prima variabile: variabile dipendente (titoli)
    #seconda variabile: variabile indipendente (indice)
    reg = sm.OLS(df_esempio[column],df_esempio['rendimento']).fit() 
    resids[column] = reg.resid
```

```python
#conversione in df dei residui
residui = pd.DataFrame(resids)
residui
```

```python
#media dei residui relativi agli ultimi 6-1 mesi
residui_media = pd.DataFrame(residui.iloc[6:11].mean()).reset_index()
```

```python
#ordinamento in senso crescente dei residui per trovare i titoli con media più alta
residui_top = residui_media.sort_values(0, ascending = False)
```

```python
#estrazione dei 10 titoli con media dei residui più alta
residui_top.head(10)
```

```python
#creazione del portafoglio
portafoglio = list(residui_top['index'].head(10))
portafoglio
```

```python
#rendimenti percentuali dei titoli del portafoglio nel mese di maggio
rendimento_portafoglio = rendimento_SP[portafoglio].loc['2019-05-31']
rendimento_portafoglio
```

```python
#calcolo rendimento medio
rendimento_portafoglio.mean()
```

```python
#24 mesi

mesi = ['maggio 19', 'giugno 19', 'luglio 19', 'agosto 19',
        'settembre 19', 'ottobre 19', 'novembre 19', 'dicembre 19', 
        'gennaio 20', 'febbrio 20', 'marzo 20', 'aprile 20', 
        'maggio 20', 'giugno 20', 'luglio 20', 'agosto 20', 
        'settembre 20', 'ottobre 20', 'novembre 20', 'dicembre 20', 
        'gennaio 21', 'febbrio 21', 'marzo 21', 'aprile 21']
```

```python
###################SP500

titoli = list(SP_rendimenti_log.columns[:-1]) #lista dei titoli escludendo l'indice
portafogli_mese_SP = {} #creazione dizionario dei portafogli mensili
rendimenti_mese_SP = {} #creazione dizionario dei rendimenti dei portafogli mensili

for i in range(24): #iterazione sui 24 mesi che costituiscono i due anni di investimenti
    
    #estrazione dei rendimenti logaritmici dei 12 mesi precedenti al mese d'investimento
    df_temp = SP_rendimenti_log[0+i : 12+i] 
    
    #regressione dei rendimenti di ogni titolo sui rendimenti dell'indice
    resids = {}
    for column in titoli:
        reg = sm.OLS(df_temp[column],df_temp['rendimento']).fit()
        resids[column] = reg.resid
        
    #conversione in dataframe dei residui
    residui = pd.DataFrame(resids)

    #media dei residui degli ultimi 6-1 mesi rispetto al mese d'investimento
    residui_media = pd.DataFrame(residui.iloc[6:11].mean()).reset_index()

    #ordinamento dei titoli con media dei residui più alta 
    residui_top = residui_media.sort_values(0, ascending = False)

    #estrazione dei primi 10 titoli che costituiranno il portafoglio
    portafogli_mese_SP[mesi[i]] = list(residui_top['index'].head(10))

    #estrazione dei rendimenti percentuali mensili dei titoli nel portafoglio in quel mese d'investimento  
    rendimenti_mese_SP[mesi[i]] = rendimento_SP[portafogli_mese_SP[mesi[i]]].iloc[i+13]
```

```python
###################EURONEXT

titoli = list(EN_rendimenti_log.columns[:-1]) #lista dei titoli escludendo l'indice
portafogli_mese_EN = {} #creazione dizionario dei portafogli mensili
rendimenti_mese_EN = {} #creazione dizionario dei rendimenti dei portafogli mensili

for i in range(24): #iterazione sui 24 mesi che costituiscono i due anni di investimenti
    
    #estrazione dei rendimenti logaritmici dei 12 mesi precedenti al mese d'investimento
    df_temp = EN_rendimenti_log[0+i : 12+i] 
    
    #regressione dei rendimenti di ogni titolo sui rendimenti dell'indice
    resids = {}
    for column in titoli:
        reg = sm.OLS(df_temp[column],df_temp['rendimento']).fit()
        resids[column] = reg.resid
        
    #conversione in dataframe dei residui
    residui = pd.DataFrame(resids)

    #media dei residui degli ultimi 6-1 mesi rispetto al mese d'investimento
    residui_media = pd.DataFrame(residui.iloc[6:11].mean()).reset_index()

    #ordinamento dei titoli con media dei residui più alta 
    residui_top = residui_media.sort_values(0, ascending = False)

    #estrazione dei primi 10 titoli che costituiranno il portafoglio
    portafogli_mese_EN[mesi[i]] = list(residui_top['index'].head(10))

    #estrazione dei rendimenti percentuali mensili dei titoli nel portafoglio in quel mese d'investimento  
    rendimenti_mese_EN[mesi[i]] = rendimento_EN[portafogli_mese_EN[mesi[i]]].iloc[i+13]

```
